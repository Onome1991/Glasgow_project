1.	Visualising Culmen Length Across Species
Create a figure that displays the mean and variance of the culmen length (the length of the penguin’s bill) for each species (Adélie, Gentoo, Chinstrap). Use appropriate visual tools (e.g., bar plots with error bars or box plots) to clearly represent the data for each species.




```python
import plotly.express as px
import plotly.graph_objects as go

# Box Plot with Mean
fig1 = px.box(
    data,
    x="species",
    y="bill_length_mm",
    color="species",
    points="all",  # Show individual points
    title="Culmen Length Across Penguin Species"
)

# Add mean points
means = data.groupby("species")["bill_length_mm"].mean().reset_index()
fig1.add_trace(go.Scatter(
    x=means["species"],
    y=means["bill_length_mm"],
    mode="markers",

    name="Mean"
))

fig1.update_layout(template="plotly_white", yaxis_title="Culmen Length (mm)")
fig1.show()

# Bar Plot with Mean and Standard Deviation
culmen_stats = data.groupby("species")["bill_length_mm"].agg(["mean", "std"]).reset_index()

fig2 = go.Figure()

fig2.add_trace(go.Bar(
    x=culmen_stats["species"],
    y=culmen_stats["mean"],
    marker=dict(color="lightblue"),
    error_y=dict(type="data", array=culmen_stats["std"], visible=True),
    name="Mean Culmen Length"
))

fig2.update_layout(
    title="Mean Culmen Length Across Penguin Species with Variance",
    xaxis_title="Species",
    yaxis_title="Mean Culmen Length (mm)",
    template="plotly_white"
)

fig2.show()
```

```python
# Select relevant numerical columns
num_features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]

# Pair Plot to visualize relationships
sns.pairplot(data, hue="species", vars=num_features, palette="Set2", markers=["o", "s", "D"])
plt.suptitle("Pair Plot: Culmen Length vs Other Features", y=1.02)
plt.show()

# Compute correlation matrix
corr_matrix = data[num_features].corr()

# Plot Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

# Add title
plt.title("Correlation Heatmap: Culmen Length vs Other Features")
plt.show()
```
